```python
from csv import reader
 
with open('Ruthenium_Prices.csv', 'r') as read_obj:
    csv_reader = reader(read_obj)
   
    tupleHeader = ("Id","Date","Price")
    dataList = []
    dataList.append(tupleHeader)
    
    endIndex = 335
    rowIndex = 0;
    for row in csv_reader:
        x = ""
        x = x.join(row)
        splittedString = x.split(';')
        if(rowIndex!=0):
            date = splittedString[0]
            price = float(splittedString[1])
            tupleA = (endIndex , date , price)
            dataList.append(tupleA)
            endIndex = endIndex -1
        rowIndex = rowIndex + 1
       
    dataList.pop(0)  #remove the headers
    dataList.reverse()   #order the list
    dataList.insert(0,tupleHeader)   #insert the header again
    #print(dataList)

    idList = []
    dateList = []
    priceList = []
    for i in range(1,336):
        tupleTemp = dataList[i]
        idList.append(dataList[i][0])
        dateList.append(dataList[i][1])
        priceList.append(dataList[i][2])
    #print(priceList)

priceArray = np.array(priceList)
maxValueIndex = np.argmax(priceArray)
minValueIndex = np.argmin(priceArray)
#print(maxValueIndex, minValueIndex)
    
plt.plot(idList, priceList) 
#plt.scatter(idList,priceList , label= "stars", color= "green", marker= "*", s=30)
plt.xlabel('Date ') 
plt.ylabel('Ruthenium price') 
plt.title('Ruthenium price vs Date"<Date represented in the graph by IDs>"') 
#plt.legend() 
plt.show()

yearList = []
monthList = []
dayList = []
for i in range(0,len(dateList)):
    date = dateList[i]
    splittedDate = date.split('-')
    yearList.append(int(splittedDate[0]))
    monthList.append(int(splittedDate[1]))
    dayList.append(int(splittedDate[2]))
#print(dateList,"all dates")
#print(yearList)
#print(monthList)
#print(dayList)

#print(len(yearList) , len(monthList) , len(dayList) , len(priceList), priceList[0] , yearList[0], monthList[0] , dayList[0])
year_train , year_test , month_train , month_test , day_train , day_test , price_train , price_test = train_test_split(yearList, monthList, dayList, priceList,test_size=0.2, random_state=0)
#print(price_test)
#year_train_array = np.array(year_train)
#year_test_array = np.array(year_test)
#month_train_array = np.array(month_train)
#month_test_array = np.array(month_test)
#day_train_array = np.array(day_train)
#day_test_array = np.array(day_test)
#price_train_array = np.array(price_train)
price_test_array = np.array(price_test)

#year_train_array = year_train_array.reshape(1,-1)
#month_train_array = month_train_array.reshape(1,-1)
#day_train_array = day_train_array.reshape(1,-1)
#price_train_array = price_train_array.reshape(1,-1)

featuresTrain = pd.DataFrame()
featuresTrain['Year Train'] = year_train
featuresTrain['Month Train'] = month_train
featuresTrain['Day Train'] = day_train
#print(featuresTrain)

priceTrain = pd.DataFrame()
priceTrain['Price Train'] = price_train
#print(priceTrain)


featuresTest = pd.DataFrame()
featuresTest['Year Test'] = year_test
featuresTest['Month Test'] = month_test
featuresTest['Day Test'] = day_test
#print(featuresTest,"Features Test")

priceTest = pd.DataFrame()
priceTest['Price Test'] = price_test
#print(priceTest)

lr = LinearRegression()  
lr.fit(featuresTrain,priceTrain)
#print(lr.intercept_)
print("Coefficients are: " , lr.coef_)
#################
#print(lr.score(featuresTrain,priceTrain) , "Hello")
#print(lr.score(featuresTest,priceTest) , "Hello")
#################
#coeff_df = pd.DataFrame(lr.coef_, featuresTrain.columns, columns=['Coefficient']) 
pricePredictions = lr.predict(featuresTest)
#print(pricePredictions.flatten())
#print(priceTest)
#print(price_test_array.flatten())
predictions = pd.DataFrame({'Actual': price_test_array.flatten(), 'Predicted': pricePredictions.flatten()})
#predictions = predictions.head(40)
print(predictions)

predictions.plot(kind='bar',figsize=(10,8))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()

#print('Mean Absolute Error:', metrics.mean_absolute_error(price_test_array, pricePredictions))  
#print('Mean Squared Error:', metrics.mean_squared_error(price_test_array, pricePredictions))  
#print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(price_test_array, pricePredictions)))

inputDate = input("Enter your future date to predict the price of Ruthenium. Date should be entered in this form yyyy-mm-dd ") 
#print(inputDate)
splittedInputDate = inputDate.split('-')
inputYear = splittedInputDate[0]
inputMonth = splittedInputDate[1]
inputDay = splittedInputDate[2]
#print(inputYear)
inputYearArr = [inputYear]
inputMonthArr = [inputMonth]
inputDayArr = [inputDay]
#print(inputYearArr)
inputDateDF = pd.DataFrame()
inputDateDF['Year input'] = inputYearArr
inputDateDF['Month input'] = inputMonthArr
inputDateDF['Day input'] = inputDayArr
#print(inputDateDF)
ouputPricePrediction = lr.predict(inputDateDF)
print("The predicted price for the input date is" , ouputPricePrediction.flatten())
```
